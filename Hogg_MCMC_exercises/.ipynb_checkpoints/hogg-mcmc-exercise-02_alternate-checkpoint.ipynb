```python
def q(x):
    if rand() > 0.5:
        # Jump to the right 
        x_prime = x + 10*rand()
    else:
        x_prime = x - 10*rand()
    return x_prime
```

```python
def f(x):
    return exp(-0.5*x*x/(10*10))
```

```python
xs = linspace(-12, 12, 100)
plot(xs, [f(x) for x in xs])
```

```python
xs = [3]
for i in range(10000):
    x = xs[-1]
    x_new = q(x)
    accept_ratio = f(x_new)/f(x)
    if rand() < accept_ratio:
        xs.append(x_new)
    else:
        xs.append(x)
```
