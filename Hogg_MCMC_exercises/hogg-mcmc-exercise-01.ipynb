```python
%pylab inline
%config InlineBackend.figure_format = 'retina'
from scipy.stats import kurtosis
from scipy.stats import skew
```

```python
# Initialize the x- and y-arrays that we will use to plot
x_axis = []
mean = []
variance = []
skew_list = []
kurt_list = []

for i in range(1,11):
    # K = 4^i
    K = 4**i
    # generate a sample from 0 < x < 1 of size K
    sample = np.random.uniform(low=0, high=1, size=K)
    
    # mean
    xbar = np.average(sample)
    # variance
    s2 = np.var(sample)
    # skewness
    b = skew(sample)
    # kurtosis
    kurt = kurtosis(sample)
    
    # Append 1/K to the x-array from earlier
    x_axis.append(1.0/K)
    # Append to the arrays of mean, variance, skewness, and kurtosis
    mean.append(xbar)
    variance.append(s2)
    skew_list.append(b)
    kurt_list.append(kurt)
    
    print(i)
    print(sample)
    
# Convert to arrays for plotting
x_axis = np.array(x_axis)
mean = np.array(mean)
variance = np.array(variance)
skew_list = np.array(skew_list)
kurt_list = np.array(kurt_list)

# Generate plots
fontsize = 8

# change x-axis to log scale

ax0 = subplot(221)
ax0.set_xscale('log')
plot(x_axis, mean-0.5)
ax0.set_title('Mean', fontsize=fontsize)

ax1 = subplot(222)
ax1.set_xscale('log')
plot(x_axis, variance)
ax1.set_title('Variance', fontsize=fontsize)

ax2 = subplot(223)
ax2.set_xscale('log')
plot(x_axis, skew_list)
ax2.set_title('Skewness', fontsize=fontsize)

ax3 = subplot(224)
ax3.set_xscale('log')
plot(x_axis, kurt_list)
ax3.set_title('Kurtosis', fontsize=fontsize)

show()
```
